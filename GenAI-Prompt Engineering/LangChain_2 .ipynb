```python
from langchain.agents import initialize_agent, Tool
from langchain_groq import ChatGroq

def calculator(input):
    try:
        return str(eval(input))
    except:
        return "Error"

tools = [
    Tool(name="Calculator", func=calculator, description="Math tool")
]

llm = ChatGroq(model="llama-3.3-70b-versatile")

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

print(agent.run("What is 20 * 5?"))
```

Hands-on Project

```python
from langchain_groq import ChatGroq
from langchain.memory import ConversationBufferMemory
from langchain.agents import initialize_agent, Tool
import os

# Set API Key (Colab)
os.environ["GROQ_API_KEY"] = "your api key"

# LLM
llm = ChatGroq(model="llama-3.3-70b-versatile")

# Memory
memory = ConversationBufferMemory()

# Tool (Calculator)
def calculator(input):
    try:
        return str(eval(input))
    except:
        return "Error"

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Performs math operations"
    )
]

# Agent
agent = initialize_agent(
    tools,
    llm,
    memory=memory,
    agent="zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True
)

# Run assistant
print(agent.run("What is 25 * 4?"))
print(agent.run("Add 10 to that"))
print(agent.run("Explain AI in simple terms"))
```

```python
#1. Customer Support Chatbot
from langchain_groq import ChatGroq
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
import os

os.environ["GROQ_API_KEY"] = "your api key"

llm = ChatGroq(model="llama-3.3-70b-versatile")

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

print(conversation.run("Hi, I need help with my order"))
print(conversation.run("Where is my order?"))
```

```python
#2. Document Question Answering (RAG)
from langchain.schema import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Documents
docs = [
    Document(page_content="AI is transforming industries."),
    Document(page_content="LangChain helps build LLM apps."),
]

# Embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Vector DB
db = FAISS.from_documents(docs, embeddings)

# Query
results = db.similarity_search("What is LangChain?")

for r in results:
    print(r.page_content)
```
